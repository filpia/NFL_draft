```python
%matplotlib inline 

import numpy as np
import pandas as pd # pandas
import matplotlib.pyplot as plt # module for plotting 
```

###Preprocessing###

I chose to value my draft picks by position according to the Weibull distribution set forth by Massey in his [paper](http://www.nber.org/papers/w11270).

```python
# val(t)= e^ (-lambda*(t-1)^Beta)
# lambda=0.148, beta=0.700
#using Weibull parameters from the Massey paper
# Uses the no. 1 overall pick as numeraire

def draft_val(pick):
    lam=0.148
    beta=0.700
    return np.exp(-1.0*lam*(pick-1)**beta)
```

```python
#create dataframes for each draft since 2008
draft_dict={}
for year in range(2008,2015):
    draft_dict[str(year)]=pd.io.parsers.read_csv("C:/Users/Filip/Desktop/Harvard/NFL draft/"
                                                 +str(year)+"_draft.csv")
    draft_dict[str(year)]['Year']=year
    ## use the Weibull distribution function defined later in the code
    draft_dict[str(year)]['Pick_val']=draft_val(draft_dict[str(year)]['Player']).tolist()


draft_master=pd.concat(draft_dict)
cols=draft_master.columns.tolist()
draft_master=draft_master.reset_index()
draft_master=draft_master[cols]
draft_master.head()
```

```python
#Create one dataframe containing the record of every team going back to 2008
# Also cols for win differential compared to previous year

records=pd.io.parsers.read_csv("C:/Users/Filip/Desktop/Harvard/NFL draft/2008_record.csv")
records=records.sort('Team')
records=records[['Team','Wins']]
records.columns=['Team','2008']

for year in range(2009,2015):
    temp=pd.io.parsers.read_csv("C:/Users/Filip/Desktop/Harvard/NFL draft/"+str(year)+"_record.csv")
    wins=temp.sort('Team')['Wins'].tolist()
    records[str(year)]=wins

# one-year differentials
records['0809_diff']=records['2009']-records['2008']
records['0910_diff']=records['2010']-records['2009']
records['1011_diff']=records['2011']-records['2010']
records['1112_diff']=records['2012']-records['2011']
records['1213_diff']=records['2013']-records['2012']
records['1314_diff']=records['2014']-records['2013']

# two-year differentials
# the col name, 'X_2yr'=- (wins in yr x)-avg( wins in yr x+1, win in yr x+2)
records['08_2yr']=(records['2009']+records['2010'])*1.0/2-records['2008']
records['09_2yr']=(records['2010']+records['2011'])*1.0/2-records['2009']
records['10_2yr']=(records['2011']+records['2012'])*1.0/2-records['2010']
records['11_2yr']=(records['2012']+records['2013'])*1.0/2-records['2011']
records['12_2yr']=(records['2013']+records['2014'])*1.0/2-records['2012']

# three-year differentials
records['08_3yr']=(records['2009']+records['2010']+records['2011'])*1.0/3-records['2008']
records['09_3yr']=(records['2012']+records['2010']+records['2011'])*1.0/3-records['2009']
records['10_3yr']=(records['2012']+records['2013']+records['2011'])*1.0/3-records['2010']
records['11_3yr']=(records['2012']+records['2013']+records['2014'])*1.0/3-records['2011']

records.head()
```

```python
#create list of teams
team_list=records['Team'].tolist()

draft_team_year=draft_master.groupby(['Year','Team'])

draft_summary={}

#create a dictionary that holds the sum of a team's draft pick in a given year
for year in range(2008,2015):
    year_temp={}
    
    for team in team_list:
        team_temp=draft_team_year.get_group((year,team))
        year_temp[team]=sum(team_temp['Pick_val'])
    draft_summary[str(year)]=year_temp
```

##Multiple Regression##

Going to run a multiple regression using two successive years of cumlative draft values as predictors to predict our value of interest, win differential. The 2-year win differential for year x is found above by averaging the wins in years x+1 and x+2 and subtracting from that value the number of wins in year x. To see the code where the win differential was computed, see the preprocessing section above.


```python

team_subset=records[records['2008']<=min_wins]
team_subset_bool=records['2008']<=min_wins
team_subset_list=team_subset['Team'].tolist()

#build lists of draft pick values and win differentials
draft_log=[]
record_log=[]

for team in team_subset_list:
    team_pair=[]
    team_pair.append(draft_summary['2009'][team])
    team_pair.append(draft_summary['2010'][team])
    draft_log.append(team_pair)
    record_log.append(records[records['Team']==team]['08_2yr'].values[0])

team_subset=records[records['2009']<=min_wins]
team_subset_bool=records['2009']<=min_wins
team_subset_list=team_subset['Team'].tolist()

for team in team_subset_list:
    team_pair=[]
    team_pair.append(draft_summary['2010'][team])
    team_pair.append(draft_summary['2011'][team])
    draft_log.append(team_pair)    
    record_log.append(records[records['Team']==team]['09_2yr'].values[0])
    
####

team_subset=records[records['2010']<=min_wins]
team_subset_bool=records['2010']<=min_wins
team_subset_list=team_subset['Team'].tolist()


for team in team_subset_list:
    team_pair=[]
    team_pair.append(draft_summary['2011'][team])
    team_pair.append(draft_summary['2012'][team])
    draft_log.append(team_pair)    
    record_log.append(records[records['Team']==team]['10_2yr'].values[0])
####

team_subset=records[records['2011']<=min_wins]
team_subset_bool=records['2011']<=min_wins
team_subset_list=team_subset['Team'].tolist()

for team in team_subset_list:
    team_pair=[]
    team_pair.append(draft_summary['2012'][team])
    team_pair.append(draft_summary['2013'][team])
    draft_log.append(team_pair)    
    record_log.append(records[records['Team']==team]['11_2yr'].values[0])####

team_subset=records[records['2012']<=min_wins]
team_subset_bool=records['2012']<=min_wins
team_subset_list=team_subset['Team'].tolist()

for team in team_subset_list:
    team_pair=[]
    team_pair.append(draft_summary['2013'][team])
    team_pair.append(draft_summary['2014'][team])
    draft_log.append(team_pair)    
    record_log.append(records[records['Team']==team]['12_2yr'].values[0])####
    
draft_df=pd.DataFrame(draft_log)
draft_df.columns=['Sum Draft val yr x+1','Sum draft val x+2']
    
draft_with_const=sm.add_constant(draft_df)
# do mult regr with a const
mult_regr_const = sm.OLS(record_log, draft_with_const).fit()

mult_regr_const.summary()
```

##Conclusions##

The structure of the annual draft looks to help bring parity to the league. The worst teams are given higher picks which have been found to have [significantly greater value](http://www.nber.org/papers/w11270) than picks later in the draft.

I sought to investigate whether one can predict how well a team will perform in coming years based solely upon the strength of their draft classes. I found that a model incorporating the value of 2 prior draft classes can account for 55% of the variation in win differentials for teams with .500-or-worse records. An additional year as a predictor did not yield an improved model, but further work can be done to verify this finding by inlcuding years prior to 2008.

While an adjusted R-squared of 54% is not near 1.00, it suggests that having a successful draft does account for improved results. One possible reason for unexplained variation is simply the variation amongst draft picks. There has been no shortage of #1 overall picks who [didn't pan out](https://usatthebiglead.files.wordpress.com/2012/03/uspw_60452201.jpg), and plenty of late-round [gems](http://media.masslive.com/patriots/photo/10631045-large.jpg). Variation amongst picks, trades, strength of schedule, and other factors aren't accounted for in the model. Nonetheless, the findings present a valid way of incorporating draft class value to explain improved performances for teams with .500-or-worse records.
